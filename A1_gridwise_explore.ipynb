```python
#Importing Package
#importing
import numpy as np
import pandas as pd
from datetime import datetime, timedelta, timezone
import matplotlib.pyplot as plt
import os
print("all libraries are loaded!")
```

```python
folder_1 = 'pit-2019' #2019 and 2020
filename = "gw-geo-pit-mar-nov-2019-000000000000.csv" #from 00 to 99
df = pd.read_csv(folder_1 + '\\' + filename)
df
```

```python
df["DateTime"] = pd.to_datetime(df['timestamp'], unit='ms' , origin='unix')
df.head()
```

```python
# Distribution of DateTime
print(df["DateTime"].describe())
ax = df["DateTime"].hist(bins = 90, figsize=(16, 4.5)) #March April May June July August September October November
ax.set_xlabel("DateTime")
ax.set_ylabel("Occurrence")
```

Ok that was overall, but now let's explore individually for a day.

```python
#lets pick a driver_id of the df
example_person = df['driver_id'][1]
df_indi = df[df['driver_id']== example_person]
df_indi
```

```python
#pick a single date
example_date = df_indi['DateTime'].dt.strftime('%Y-%m-%d')
example_date = example_date[1]
print(example_date)

# filter by single day
df_indi = df_indi[(df_indi['DateTime'].dt.date.astype(str) == example_date)]
df_indi
```

```python
# To Create Directory If Not Exist
# https://appdividend.com/2021/07/03/how-to-create-directory-if-not-exist-in-python/
path = folder_1+'\\'+example_date
os.makedirs(path, exist_ok=False)
print("The new directory is created!")

#export
df_indi.to_csv(path+ "\\" + example_person + ".csv")
```
